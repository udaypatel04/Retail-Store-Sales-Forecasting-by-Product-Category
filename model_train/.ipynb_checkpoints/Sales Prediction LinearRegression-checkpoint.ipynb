## Filling missing values
### Outlet_Type (filling with mode)

```python
df['Outlet_Size'].isnull().sum()
```

```python
df['Outlet_Size'].value_counts()
```

```python
df['Outlet_Size'].mode()
```

```python
mode_of_outlet_size=df.pivot_table(values='Outlet_Size',columns='Outlet_Type', aggfunc= lambda x:x.mode()[0])
```

```python
mode_of_outlet_size
```

```python
miss_values=df['Outlet_Size'].isnull()
```

```python
miss_values.value_counts()
```

```python
df.loc[miss_values,'Outlet_Size']=df.loc[miss_values,'Outlet_Type'].apply(lambda x: mode_of_outlet_size[x])
```

## Label Enconding

```python
encoder=LabelEncoder()
df['Item_Identifier']=encoder.fit_transform(df['Item_Identifier'])
df['Item_Fat_Content']=encoder.fit_transform(df['Item_Fat_Content'])
df['Item_Type']=encoder.fit_transform(df['Item_Type'])
df['Outlet_Identifier']=encoder.fit_transform(df['Outlet_Identifier'])
df['Outlet_Size']=encoder.fit_transform(df['Outlet_Size'])
df['Outlet_Location_Type']=encoder.fit_transform(df['Outlet_Location_Type'])
df['Outlet_Type']=encoder.fit_transform(df['Outlet_Type'])
```

```python
df.head()
```

## Splitting Data Into Test and Train Parts

```python
X= df.drop(columns='Item_Outlet_Sales',axis=1)
Y=df['Item_Outlet_Sales']
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
df.sample(6)
```

## Machine Learning Model Training

### 3. LinearRegression

```python
from sklearn.linear_model import LinearRegression
linearRegressionModel = LinearRegression()
linearRegressionModel.fit(X_train, Y_train)
y_pred_lr = linearRegressionModel.predict(X_test)

# Calculate R² score
lr_score = r2_score(Y_test, y_pred_lr)
print(f"Linear Regression R² score on test set: {lr_score * 100:.2f}%")
```

```python

X = df[['Item_MRP']]
y = df['Item_Outlet_Sales']

print("Training the Linear Regression model...")
model = LinearRegression()
model.fit(X, y)
print("Model training complete.")

print("Generating the linear regression graph...")

plt.figure(figsize=(10, 6))

# Plot the actual data points as a scatter plot
plt.scatter(X, y, color='skyblue', label='Actual Sales Data')

# Plot the regression line based on the model's predictions
# The model's predictions form a straight line
plt.plot(X, model.predict(X), color='red', linewidth=3, label='Linear Regression Line')

# Add labels and a title for clarity
plt.title('Linear Regression: Item_MRP vs. Item_Outlet_Sales', fontsize=16)
plt.xlabel('Item Maximum Retail Price (Item_MRP)', fontsize=12)
plt.ylabel('Item Outlet Sales (Item_Outlet_Sales)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Save the plot to a file
output_file = "plots/linear_regression_graph.png"
plt.savefig(output_file)
print(f"Graph saved successfully as '{output_file}'.")

# You can also use plt.show() if you're in an interactive environment
# plt.show()
```
